```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Настройка стиля
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✅ Библиотеки загружены")
```

```python
# Загружаем данные
customers = pd.read_csv('../data/customers.csv')
dishes = pd.read_csv('../data/dishes.csv')
categories = pd.read_csv('../data/categories.csv')
reservations = pd.read_csv('../data/reservations.csv')
events = pd.read_csv('../data/events.csv')

print(f"Клиенты: {len(customers)}")
print(f"Блюда: {len(dishes)}")
print(f"Категории: {len(categories)}")
print(f"Бронирования: {len(reservations)}")
print(f"Мероприятия: {len(events)}")
```

```python
# Анализ клиентов
customers.head()
customers.describe()
customers['bonuses'].hist(bins=20)
plt.title('Распределение бонусов у клиентов')
plt.xlabel('Бонусы')
plt.ylabel('Количество клиентов')
plt.show()
```

```python
# Анализ блюд
dishes.head()

# Топ-5 дорогих блюд
dishes.nlargest(5, 'price')[['name', 'price']]

# Топ-5 дешёвых блюд
dishes.nsmallest(5, 'price')[['name', 'price']]

# Распределение цен
dishes['price'].hist(bins=15)
plt.title('Распределение цен на блюда')
plt.xlabel('Цена (руб.)')
plt.ylabel('Количество блюд')
plt.show()
```

```python
# Анализ бронирований
reservations.head()

# Количество бронирований по месяцам
reservations['reservation_date'] = pd.to_datetime(reservations['reservation_date'])
reservations['month'] = reservations['reservation_date'].dt.month
reservations['month'].value_counts().sort_index().plot(kind='bar')
plt.title('Бронирования по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество')
plt.show()

# Статусы бронирований
reservations['status'].value_counts()
```
